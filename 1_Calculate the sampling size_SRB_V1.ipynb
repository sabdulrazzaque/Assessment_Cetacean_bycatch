```python
import math
from scipy.stats import norm

def calculate_proportion_confidence_interval(sample_size, successes, confidence_level):
    # Calculate the sample proportion
    sample_proportion = successes / sample_size

    # Calculate the standard error using the normal distribution approximation (Wald interval)
    standard_error_wald = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

    # Calculate the z-score for the desired confidence level
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)

    # Calculate the confidence interval using the normal distribution approximation (Wald interval)
    confidence_interval_wald = (
        sample_proportion - z_score * standard_error_wald,
        sample_proportion + z_score * standard_error_wald
    )

    # Calculate the confidence interval using the binomial distribution
    standard_error_binomial = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)
    confidence_interval_binomial = (
        sample_proportion - z_score * standard_error_binomial,
        sample_proportion + z_score * standard_error_binomial
    )

    # Calculate the confidence interval using the Wilson score interval
    z_squared = z_score ** 2
    denominator = 1 + (z_squared / sample_size)
    adjusted_proportion = sample_proportion + (z_squared / (2 * sample_size))
    adjusted_standard_error = math.sqrt((sample_proportion * (1 - sample_proportion) + (z_squared / (4 * sample_size))) / sample_size)
    confidence_interval_wilson = (
        (adjusted_proportion - (z_score * adjusted_standard_error)) / denominator,
        (adjusted_proportion + (z_score * adjusted_standard_error)) / denominator
    )

    # Return the calculated confidence intervals
    return {
        'Wald Interval': confidence_interval_wald,
        'Binomial Interval': confidence_interval_binomial,
        'Wilson Score Interval': confidence_interval_wilson
    }

# Example usage
sample_size = 150
successes = 70
confidence_level = 0.95

confidence_intervals = calculate_proportion_confidence_interval(sample_size, successes, confidence_level)

# Display the results
for interval_name, interval in confidence_intervals.items():
    print(f"{interval_name}: {interval}")

```

```python
import math
from scipy.stats import norm

def proportion_confidence_interval_wald(successes, trials, confidence=0.95):
    p_hat = successes / trials
    z = norm.ppf(1 - (1 - confidence) / 2)
    margin_of_error = z * math.sqrt((p_hat * (1 - p_hat)) / trials)
    lower_bound = p_hat - margin_of_error
    upper_bound = p_hat + margin_of_error
    return lower_bound, upper_bound

def proportion_confidence_interval_binomial(successes, trials, confidence=0.95):
    p_hat = successes / trials
    z = norm.ppf(1 - (1 - confidence) / 2)
    margin_of_error = z * math.sqrt((p_hat * (1 - p_hat)) / trials + (z**2) / (4 * trials**2))
    lower_bound = p_hat - margin_of_error
    upper_bound = p_hat + margin_of_error
    return lower_bound, upper_bound

def proportion_confidence_interval_wilson(successes, trials, confidence=0.95):
    p_hat = successes / trials
    z = norm.ppf(1 - (1 - confidence) / 2)
    denominator = 1 + (z**2) / trials
    center_adjusted_probability = p_hat + (z**2) / (2 * trials)
    margin_of_error = z * math.sqrt((p_hat * (1 - p_hat) + (z**2) / (4 * trials)) / trials)
    lower_bound = (center_adjusted_probability - margin_of_error) / denominator
    upper_bound = (center_adjusted_probability + margin_of_error) / denominator
    return lower_bound, upper_bound

# Example usage
successes = 70
trials = 150
confidence = 0.95

wald_interval = proportion_confidence_interval_wald(successes, trials, confidence)
binomial_interval = proportion_confidence_interval_binomial(successes, trials, confidence)
wilson_interval = proportion_confidence_interval_wilson(successes, trials, confidence)

print("Wald Interval:")
print(f"Lower Bound: {wald_interval[0]}")
print(f"Upper Bound: {wald_interval[1]}")
print()
print("Binomial Interval:")
print(f"Lower Bound: {binomial_interval[0]}")
print(f"Upper Bound: {binomial_interval[1]}")
print()
print("Wilson Interval:")
print(f"Lower Bound: {wilson_interval[0]}")
print(f"Upper Bound: {wilson_interval[1]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

def wald_interval(x, n, alpha):
    p_hat = x / n
    z = np.abs(norm.ppf(alpha/2))
    lower = p_hat - z * np.sqrt(p_hat*(1-p_hat)/n)
    upper = p_hat + z * np.sqrt(p_hat*(1-p_hat)/n)
    return lower, upper

def binomial_interval(x, n, alpha):
    lower = binom.ppf(alpha/2, n, x/n)
    upper = binom.ppf(1-alpha/2, n, x/n)
    return lower/n, upper/n

def wilson_interval(x, n, alpha):
    p_hat = x / n
    z = np.abs(norm.ppf(alpha/2))
    denominator = 1 + z**2 / n
    center_adjusted = p_hat + z**2 / (2 * n)
    adjusted_sqrt = np.sqrt(z**2 / n * p_hat * (1 - p_hat) / n + z**2 / (4 * n**2))
    lower = (center_adjusted - adjusted_sqrt) / denominator
    upper = (center_adjusted + adjusted_sqrt) / denominator
    return lower, upper

def clopper_pearson_interval(x, n, alpha):
    lower = binom.ppf(alpha/2, n, x/n)
    upper = binom.ppf(1-alpha/2, n, x/n+1e-9)  # Add a small value to avoid upper bound being exactly 1
    return lower/n, upper/n

def plot_intervals(intervals, labels):
    x = np.arange(len(intervals))
    plt.figure(figsize=(10, 6))
    for i, interval in enumerate(intervals):
        plt.errorbar(x[i], interval[0], yerr=[[interval[0]], [interval[1]]], marker='o', label=labels[i])
    plt.xticks(x, labels)
    plt.ylim(0, 1)
    plt.xlabel('Interval Method')
    plt.ylabel('Proportion')
    plt.title('Proportion Confidence Intervals')
    plt.legend()
    plt.show()

# Parameters
x = 75  # Number of successes
n = 150  # Total number of trials
alpha = 0.05  # Significance level

# Calculate confidence intervals
wald = wald_interval(x, n, alpha)
binomial = binomial_interval(x, n, alpha)
wilson = wilson_interval(x, n, alpha)
clopper_pearson = clopper_pearson_interval(x, n, alpha)

# Display the scores and results
print("Scores and Results:")
print(f"Number of successes (x): {x}")
print(f"Total number of trials (n): {n}")
print(f"Significance level (alpha): {alpha}")

print("\nConfidence Intervals:")
print("Wald Interval:", wald)
print("Binomial Interval:", binomial)
print("Wilson Interval:", wilson)
print("Clopper-Pearson Interval:", clopper_pearson)


# Plot the intervals
intervals = [wald, binomial, wilson, clopper_pearson]
labels = ['Wald', 'Binomial', 'Wilson', 'Clopper-Pearson']
plot_intervals(intervals, labels)

# Recommendation based on interval width
widths = [interval[1] - interval[0] for interval in intervals]
best_method = labels[np.argmin(widths)]
print(f"\nThe recommended method is: {best_method}")
```

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alternative='two-sided'):
    # Calculate the standard error
    std_error = population_std / np.sqrt(sample_size)

    # Calculate the test statistic
    z_score = (sample_mean - population_mean) / std_error

    # Calculate the p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
    elif alternative == 'less':
        p_value = norm.cdf(z_score)

    # Return the test statistic and p-value
    return z_score, p_value

# Parameters
sample_mean = 5.2  # Sample mean
population_mean = 5.0  # Population mean
population_std = 1.5  # Population standard deviation
sample_size = 150  # Sample size
alternative = 'two-sided'  # Alternative hypothesis ('two-sided', 'greater', 'less')

# Perform the Z-test
z_score, p_value = z_test(sample_mean, population_mean, population_std, sample_size, alternative)

# Display the results
print("Z-test Results:")
print(f"Sample Mean: {sample_mean}")
print(f"Population Mean: {population_mean}")
print(f"Population Standard Deviation: {population_std}")
print(f"Sample Size: {sample_size}")
print(f"Alternative Hypothesis: {alternative}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def sample_size_z_test(x, n, p, alpha):
    p_hat = x / n
    z_critical = norm.ppf(1 - alpha/2)
    standard_error = np.sqrt(p * (1 - p) / n)
    margin_of_error = z_critical * standard_error
    lower_bound = p_hat - margin_of_error
    upper_bound = p_hat + margin_of_error
    return lower_bound, upper_bound

# Parameters
x = 75  # Number of successes
n = 150  # Sample size
p = 0.5  # Assumed population proportion
alpha = 0.05  # Significance level

# Perform sample size Z-test
lower_bound, upper_bound = sample_size_z_test(x, n, p, alpha)

# Print the scores and results
print("Scores and Results:")
print(f"Number of successes (x): {x}")
print(f"Sample size (n): {n}")
print(f"Assumed population proportion (p): {p}")
print(f"Significance level (alpha): {alpha}")
print("")

print("Sample Size Z-Test:")
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Plot the scores
scores = [lower_bound, upper_bound]
labels = ["Lower Bound", "Upper Bound"]
plt.bar(labels, scores)
plt.ylim(0, 1)
plt.xlabel('Interval')
plt.ylabel('Proportion')
plt.title('Sample Size Z-Test')
plt.show()
```
