```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset into a pandas DataFrame
df = pd.read_csv('Survey_F2F_Cetacean.csv')

# Convert 'Time_Start' and 'Time_End' columns to datetime
df['Time_Start'] = pd.to_datetime(df['Time_Start'])
df['Time_End'] = pd.to_datetime(df['Time_End'])

# Calculate the time difference in minutes
df['Time_Difference'] = (df['Time_End'] - df['Time_Start']).dt.total_seconds() / 60

# Assign serial numbers to the time differences based on 'Recorder'
df['Serial_Number_Recorder'] = df.groupby('Recorder').cumcount() % 52 + 1

# Assign serial numbers to the time differences based on 'District'
df['Serial_Number_District'] = df.groupby('District').cumcount() % 52 + 1

# Calculate the results by Recorder and Serial Number
results_recorder = df.groupby(['Recorder', 'Serial_Number_Recorder'])['Time_Difference'].mean()

# Calculate the results by District and Serial Number
results_district = df.groupby(['District', 'Serial_Number_District'])['Time_Difference'].mean()

# Print the results
print(results_recorder)
print(results_district)

# Display the results
display(results_recorder)
display(results_district)

# Plot the results by Recorder
plt.figure(figsize=(10, 5))
for recorder in df['Recorder'].unique():
    plt.plot(
        df[(df['Recorder'] == recorder) & (df['Serial_Number_Recorder'] <= 52)]['Serial_Number_Recorder'],
        df[(df['Recorder'] == recorder) & (df['Serial_Number_Recorder'] <= 52)]['Time_Difference'],
        label=recorder
    )
plt.xlabel('Serial Number (Recorder)')
plt.ylabel('Time Difference (minutes)')
plt.title('Time Difference between Start and End by Recorder')
plt.legend()
plt.show()

# Plot the results by District
plt.figure(figsize=(10, 5))
for district in df['District'].unique():
    plt.plot(
        df[(df['District'] == district) & (df['Serial_Number_District'] <= 52)]['Serial_Number_District'],
        df[(df['District'] == district) & (df['Serial_Number_District'] <= 52)]['Time_Difference'],
        label=district
    )
plt.xlabel('Serial Number (District)')
plt.ylabel('Time Difference (minutes)')
plt.title('Time Difference between Start and End by District')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset into a pandas DataFrame
df = pd.read_csv('Survey_F2F_Cetacean.csv')

# Convert 'Time_Start' and 'Time_End' columns to datetime
df['Time_Start'] = pd.to_datetime(df['Time_Start'])
df['Time_End'] = pd.to_datetime(df['Time_End'])

# Calculate the time difference in minutes
df['Time_Difference'] = (df['Time_End'] - df['Time_Start']).dt.total_seconds() / 60

# Assign serial numbers to the time differences based on 'Recorder' and 'District'
df['Serial_Number_Recorder'] = df.groupby('Recorder').cumcount() % 52 + 1
df['Serial_Number_District'] = df.groupby('District').cumcount() % 52 + 1

# Calculate the results by Recorder and Serial Number
results_recorder = df.groupby(['Recorder', 'Serial_Number_Recorder'])['Time_Difference'].mean()
results_recorder_std = df.groupby(['Recorder', 'Serial_Number_Recorder'])['Time_Difference'].std()

# Calculate the results by District and Serial Number
results_district = df.groupby(['District', 'Serial_Number_District'])['Time_Difference'].mean()
results_district_std = df.groupby(['District', 'Serial_Number_District'])['Time_Difference'].std()

# Print the results
print(results_recorder)
print(results_district)

# Display the results in a table
results_recorder_table = pd.DataFrame({'Average Time Difference': results_recorder, 'Standard Deviation': results_recorder_std})
results_district_table = pd.DataFrame({'Average Time Difference': results_district, 'Standard Deviation': results_district_std})
print("Results by Recorder:")
display(results_recorder_table)
print("Results by District:")
display(results_district_table)

# Plot the results by Recorder
plt.figure(figsize=(10, 5))
for recorder in df['Recorder'].unique():
    avg_values = []
    std_values = []
    for serial_number in range(1, 53):
        data = df[(df['Recorder'] == recorder) & (df['Serial_Number_Recorder'] == serial_number)]
        if not data.empty:
            avg_values.append(data['Time_Difference'].mean())
            std = data['Time_Difference'].std()
            if np.isnan(std):
                std_values.append(0)  # Replace non-finite values with 0
            else:
                std_values.append(std)
    plt.errorbar(
        range(1, len(avg_values) + 1),
        avg_values,
        yerr=std_values,
        marker='o',
        label=recorder
    )
plt.xlabel('Serial Number (Recorder)')
plt.ylabel('Time Difference (minutes)')
plt.title('Average +/- Time Difference between Start and End by Recorder')
plt.legend()
plt.show()

# Plot the results by District
plt.figure(figsize=(10, 5))
for district in df['District'].unique():
    avg_values = []
    std_values = []
    for serial_number in range(1, 53):
        data = df[(df['District'] == district) & (df['Serial_Number_District'] == serial_number)]
        if not data.empty:
            avg_values.append(data['Time_Difference'].mean())
            std = data['Time_Difference'].std()
            if np.isnan(std):
                std_values.append(0)  # Replace non-finite values with 0
            else:
                std_values.append(std)
    plt.errorbar(
        range(1, len(avg_values) + 1),
        avg_values,
        yerr=std_values,
        marker='o',
        label=district
    )
plt.xlabel('Serial Number (District)')
plt.ylabel('Time Difference (minutes)')
plt.title('Average +/- Time Difference between Start and End by District')
plt.legend()
plt.show()
```
